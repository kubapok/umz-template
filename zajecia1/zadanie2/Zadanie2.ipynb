```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flats = pd.read_csv('train/train.tsv', sep = '\t', 
                   names = ['price', 'isNew','rooms', 'floor', 'location', 'sqrMetres'])
flats_filter = flats[(flats['price'] < 1500000) & (flats.price > 150000)]
flats.price.mean()
```

```python
flats_filter.price.mean()
```

```python
from sklearn import linear_model
#create linear regression model
reg = linear_model.LinearRegression()
#train the model
X = flats_filter['sqrMetres']
X = X.values.reshape(-1,1)
Y = flats_filter['price']
#reg.fit(pd.DataFrame(flats, columns=['sqrMetres']), flats['price'])
#make predictions
model = reg.fit(X, Y)
#predict_price = reg.predict(2)
a = model.coef_[0]
b = model.intercept_
```

```python
print("Coefficient: {0}, intercept: {1}".format(a,b))
```

```python
plt.scatter(X,Y, color='orange')
plt.plot([0,200],[b,a*200+b], 'r')
plt.title('Linear regression')
plt.xlabel('sqrMetres')
plt.ylabel('price')
```
